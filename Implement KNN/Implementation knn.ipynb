The KNN or K-nearest neighbors algorithm works similar to the saying "tell me who you hang out with and I'll tell you who you are." This means that the algorithm makes its prediction based on what the majority of variables around the set of test say.

**Pseudo code of the KNN algorithm**

```python
#Function calculate_distance(point1, point2):
     #Calculate the distance between point1 and point2
     #Return the distance

#Function find_k_closest_neighbors:
     #Initialize an empty list of neighbors
     #For each point in training_data:
         #Calculate the distance between test_point and current point
         #Add (distance, current_point) to neighbor list
     #Sort the list of neighbors by distance in ascending order
     #Take the first K elements of the list
     #Return the list of K neighbors

#Function predict_class(test_points, training_data, training_labels, K):
     #For each test_point in test_points:
         #Find the K nearest neighbors using find_k_closest_neighbors
         #Count nearest neighbors tags
         #Assign the most common label as the predicted label for the test point
     #Return predicted labels
```

For show how this code works, I will use the numpy and matplotlib libreries

```python
import numpy as np
import matplotlib.pyplot as plt
```

In the next celd I create some points for a data set and i assigned it some labels, 0 or 1 and I create a data test for prove the algorithm, also I need to define the K number of neighbors that the algorithm will take in consideration for the predicion.

```python
# Crear un conjunto de datos de ejemplo
X = np.array([[1, 2], [2, 3], [3, 4], [5, 5], [6, 7]])
y = np.array([0, 0, 1, 1, 1])  # Etiquetas de clase (0: Morado, 1: Amarillo)

# Punto de prueba
x_test = np.array([4, 4])

# Definir el valor de K
K = 3
```

After created my data set I use this formule : D(p, q) = √((p1 - q1)^2 + (p2 - q2)^2 + ... + (pn - qn)^2), to find the nearest neighbors to my test point.

```python
# Calcular distancias entre el punto de prueba y los puntos de entrenamiento
distances = np.sqrt(np.sum((X - x_test) ** 2, axis=1))
```

Now that I have all the distances between the test point and the train points, I use the function argsort to order the distances from the lowest to highest and in this way I can select the "k" nearest neighbords.

```python
# Obtener los índices de los K vecinos más cercanos
nearest_indices = np.argsort(distances)[:K]
```

Finally I can observe the lebels of my nearest neighbors for be able to realize my predictions.

```python
# Contar las etiquetas de clase de los vecinos más cercanos
nearest_labels = y[nearest_indices]

# Determinar la etiqueta de clase predicha
predicted_label = np.bincount(nearest_labels).argmax()
```

In the graph bellow it is seemed in color purple the 0 lebel and in yellow the 1 lebel, the red X is my test  point and the other ones are my train points, the lines are the distances between my test point and my nearest neighbord, as it can be seemed the major of the neighbors are Yellow so my prediction will be thar my test point will be yellow to or will be a 1 lebel

```python
# Crear un gráfico para visualizar los puntos y el punto de prueba
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], c='purple', marker='o', s=100, label='label 0')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], c='yellow', marker='s', s=100, label='label 1')
plt.scatter(x_test[0], x_test[1], c='red', marker='x', s=100, label='test point')
plt.xlabel('Characteristic 1')
plt.ylabel('Characteristic 2')
plt.legend()

# Resaltar los vecinos más cercanos
for idx in nearest_indices:
    plt.plot([X[idx][0], x_test[0]], [X[idx][1], x_test[1]], 'k--', linewidth=1)

plt.title(f'Clasification with KNN (K={K}), label predicted : {predicted_label}')
plt.show()

print(f'The trainning point will be part of the label: {predicted_label}')



```

**Loss function and Optimization function**

Loss Functions:

A loss function, also known as a cost function or error function, is a measure used to quantify how well or how poorly a machine learning model is making predictions compared to the ground truth labels in the set. of data. In other words, the loss function evaluates how close or far the model's predictions are from the actual values it is trying to predict.
The main goal of a loss function is to provide a numerical measure that the optimization algorithm can minimize during model training. The smaller the value of the loss function, the better the fit of the model to the training data.
Examples of common loss functions include mean square error (MSE) for regression problems, cross-entropy for classification problems, and several other functions designed for specific tasks.

Optimization Functions:

Optimization functions are used to tune or train a machine learning model in a way that minimizes or maximizes an objective function. In the context of machine learning, optimization functions are used to find the model parameters that minimize the loss function.
The process of training a machine learning model typically involves iteration of parameter tuning. During each iteration, the optimization function adjusts the model parameters so that the loss function decreases, leading to better model performance on the training data.
Examples of optimization functions include gradient descent, which is a widely used algorithm for optimizing machine learning models, and its variants such as stochastic gradient descent (SGD), Adam, RMSprop, among others. others.
